```python
import numpy as np 
from sklearn.datasets import make_regression
```

```python
X, y = make_regression(n_samples=4, n_features=1, n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
import matplotlib.pyplot as plt 
plt.scatter(X,y)
```

```python
# Let's apply OLS (Ordinary Least Squares)
from sklearn.linear_model import LinearRegression

reg = LinearRegression()

reg.fit(X,y)
```

```python
reg.coef_
```

```python
reg.intercept_
```

```python
plt.scatter(X,y)

plt.plot(X, reg.predict(X), color="red")
```

### Let's Apply Gradient Descent

```python
# Let's apply gradient descent assuming slope is constant m = 78.35
# and let's assume the starting value for intercept b = 0 
m = 78.35
b= 0
y_pred = ((m * X) + b ).reshape(4)
```

```python
y_pred
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X,y_pred, color="#00a65a", label='b_0=0')
plt.legend()
plt.show()
```

```python
m = 78.35 
b_0 = 0 
 

loss_slope = -2 * np.sum(y - m * X.ravel() - b_0)
loss_slope
```

```python
# Let's set the learning rate to 0.1

lr = 0.1 

# for your kind information lr * slope is known as step size
step_size = loss_slope * lr 
step_size
```

```python
# Calculating the new intercept
b_1 = b_0 - step_size
b_1
```

```python
y_pred1 = ((78.35 * X ) + b_1).reshape(4)
```

```python
y_pred1
```

```python
plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred1, color="#00a65a", label=f"b_1={b_1}")
plt.plot(X, y_pred, color="#A3E407", label="b=0")
plt.legend()
plt.show()
```

```python
# Iteration 2
loss_slope = -2 * np.sum(y - m * X.ravel() - b_1)
loss_slope
```

```python
step_size = loss_slope * lr 
step_size
```

```python
b_2 = b_1 - step_size
b_2
```

```python
y_pred2 = ((78.35 * X) + b_2 ).reshape(4)

plt.scatter(X,y)
plt.plot(X, reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred2, color="#00a65a", label=f"b_2={b_2}")
plt.plot(X,y_pred1, color="#A3E4D7", label=f"b_1={b_1}")
plt.plot(X, y_pred, color="blue", label="b_0=0")
plt.legend()
plt.show()
```

```python
# Iteration 3
loss_slope = -2 * np.sum(y - m*X.ravel() - b_2)
loss_slope
```

```python
step_size = loss_slope * lr 
step_size
```

```python
b_3 = b_2 - step_size
b_3
```

```python
y_pred3 = ((78.35 * X) + b_3).reshape(4)

plt.figure(figsize=(10,10))
plt.scatter(X,y)
plt.plot(X,reg.predict(X), color="red", label="OLS")
plt.plot(X, y_pred3, color="#00a65a", label=f"b_3={b_3}")
plt.plot(X, y_pred2, color="#A3E4D7", label=f"b_2={b_2}")
plt.plot(X, y_pred1, color="#A3E4D7", label=f"b_1={b_1}")
plt.plot(X, y_pred, color="blue", label=f"b=0")

plt.legend()
plt.show()
```

```python
b = -100 # Taking different b values for experimentation
m = 78.35 
lr = 0.1 

epochs = 10 


for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)
    b = b - (lr * loss_slope)
    y_pred = m * X + b 
    plt.plot(X, y_pred)

plt.scatter(X,y)

```

## Smaller Learning Rate: Too Slow 

```python
b = -100 
m = 78.35 
lr = 0.01 # Smaller Learning Rate 

epochs = 10 


for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)
    b = b - (lr * loss_slope)
    y_pred = m * X + b 
    plt.plot(X, y_pred)

plt.scatter(X,y)

```

```python
# You're not able to reach the actual answer because of smaller learning rate, but increment in epochs can be a solution. 
```

```python
b = -100
m = 78.35 
lr = 0.01 # Smaller Learning Rate with more epochs 

epochs = 100 # previously 10 now 100


for i in range(epochs):
    loss_slope = -2 * np.sum(y - m * X.ravel() - b)
    b = b - (lr * loss_slope)
    y_pred = m * X + b 
    plt.plot(X, y_pred)

plt.scatter(X,y)

```

```python
# Means even after smaller learning rate with high epochs you're still getting the solution. 
# But in constrast the bigger learning rate will create a zigzag pattern
```
